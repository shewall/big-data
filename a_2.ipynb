```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("Trips_Full_Data.csv")
week_count = df['Week of Date'].nunique()
    
   
distance_columns = ['Trips <1 Mile', 'Trips 1-3 Miles', 'Trips 3-5 Miles', 'Trips 5-10 Miles',
                        'Trips 10-25 Miles', 'Trips 25-50 Miles', 'Trips 50-100 Miles', 
                        'Trips 100-250 Miles', 'Trips 250-500 Miles', 'Trips 500+ Miles']
mean_trips_per_distance = df[distance_columns].mean()
average_trips_per_distance = df.groupby('Week of Date')[distance_columns].mean()
print(average_trips_per_distance)
 
 
plt.figure(figsize=(10, 6))
mean_trips_per_distance.plot(kind='bar')
plt.title('People Traveling vs Distance')
plt.xlabel('Distance Category')
plt.ylabel('Number of People Traveling')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
   

```
