```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


data = pd.read_csv('Trips_Full_Data.csv')


distance_categories = [
    'Trips <1 Mile', 'Trips 1-3 Miles', 'Trips 3-5 Miles', 'Trips 5-10 Miles',
    'Trips 10-25 Miles', 'Trips 25-50 Miles', 'Trips 50-100 Miles', 
    'Trips 100-250 Miles', 'Trips 250-500 Miles', 'Trips 500+ Miles'
]
totals = data[distance_categories].sum()


plt.figure(figsize=(12, 8))
bars = plt.bar(totals.index, totals.values, color=plt.cm.Paired(np.arange(len(totals))))

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{int(yval):,}', ha='center', va='bottom', fontsize=10, rotation=0)

plt.title('Number of Travelers by Distance Category', fontsize=16)
plt.xlabel('Distance Categories', fontsize=14)
plt.ylabel('Number of Travelers', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
