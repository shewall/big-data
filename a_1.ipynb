```python

import pandas as pd
import matplotlib.pyplot as plt


distance = pd.read_csv('Trips_by_Distance.csv')

week_count = distance['Week'].nunique()


population_staying_home = distance['Population Staying at Home'].mean()

avg_population_staying_home = distance.groupby('Week')['Population Staying at Home'].mean()


print(f"Unique weeks in dataset: {week_count}")

print("Average Population Staying at Home per Week:")
print(population_staying_home)

print("Average Population Staying at Home per Week:")
print(avg_population_staying_home)


fig = plt.figure(figsize=(15, 8))


plt.bar(avg_population_staying_home.index, avg_population_staying_home, color='orange', width=0.4)
plt.xlabel("Week")
plt.xticks(range(0, avg_population_staying_home.index.nunique(), 5), rotation=90) 

plt.ylabel("Average Population Staying at Home")
plt.title("Average Population Staying at Home per Week")

plt.grid()
plt.rcParams.update({'text.color': "black",
                     'axes.labelcolor': "black",
                     'xtick.color': "black",
                     'ytick.color': "black",
                     'font.size': 13})

plt.show()
```
