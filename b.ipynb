```python

import pandas as pd
import matplotlib.pyplot as plt

data_distance = pd.read_csv('Trips_by_Distance.csv')


data_distance = data_distance[['Date', 'Number of Trips 10-25', 'Number of Trips 50-100']]


data_distance['Date'] = pd.to_datetime(data_distance['Date'])


over_10m_10_25 = data_distance[data_distance['Number of Trips 10-25'] > 10000000]

over_10m_50_100 = data_distance[data_distance['Number of Trips 50-100'] > 10000000]


plt.figure(figsize=(10, 6))
plt.scatter(over_10m_10_25 ['Date'], over_10m_10_25 ['Number of Trips 10-25'], color='blue', label='Trips 10-25')
plt.title('Number of Trips 10-25 Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Trips 10-25')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(over_10m_50_100['Date'], over_10m_50_100['Number of Trips 50-100'], color='red', label='Trips 50-100')
plt.title('Number of Trips 50-100 Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Trips 50-100')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



```

```python

common_dates = data_distance[(data_distance['Number of Trips 10-25'] > 10000000) & (data_distance['Number of Trips 50-100'] > 10000000)]

plt.figure(figsize=(10, 6))
plt.scatter(common_dates['Date'], common_dates['Number of Trips 10-25'], color='blue', label='Trips 10-25 > 10M', alpha=0.7)
plt.scatter(common_dates['Date'], common_dates['Number of Trips 50-100'], color='red', label='Trips 50-100 > 10M', alpha=0.7)
plt.title('Direct Comparison on Common Dates')
plt.xlabel('Date')
plt.ylabel('Number of Trips')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
